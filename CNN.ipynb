# CONVOLUTION

```python
import numpy as np
def conv_1D(signal, kernel):
    
    n = len(signal)
    k = len(kernel)
    
    pad_size = k//2
    padded_signal = [0]*pad_size + signal + [0]*pad_size
    
    output = np.zeros(n)
    
    kernel = np.array(kernel)
    padded_signal = np.array(padded_signal)
    
    for i in range(n):
        print(i)
        output[i] = np.sum(kernel*padded_signal[i:i+k])
    return output

signal = [1, 2, 3, 4, 5, 6]
kernel = [1, 0, -1]
output = conv_1D(signal, kernel)
print(output)
        
        
```

```python
import numpy as np

def conv_3D(image, kernel):
    (image_h, image_w, image_ch) = image.shape
    (kernel_h, kernel_w, input_ch, output_ch) = kernel.shape
    
    pad_h = kernel_h // 2
    pad_w = kernel_w // 2
    
    padded_image = np.pad(image, pad_width=((pad_h, pad_h), (pad_w, pad_w), (0, 0)), mode='constant', constant_values=0)
    output = np.zeros((image_h, image_w, output_ch))
    
    for i in range(image_h):
        for j in range(image_w):
            for k in range(output_ch):
                output[i, j, k] = np.sum(kernel[:, :, :, k] * padded_image[i:i+kernel_h, j:j+kernel_w, :])
    
    return output

# Example usage with properly shaped image and kernel
image = np.random.randn(5, 5, 3)  # Example input image with height=5, width=5, channels=3
kernel = np.random.randn(3, 3, 3, 2)  # Example 3x3 kernel for 3 input channels and 2 output channels

# Perform the convolution operation
output = conv_3D(image, kernel)

print('Input image shape:', image.shape)
print('Kernel shape:', kernel.shape)
print('Output shape:', output.shape)

print('\nInput image:')
print(image)

print('\nKernel:')
print(kernel)

print('\nOutput:')
print(output)
```
